```python
folder_path=r'C:\Users\wojjj\OneDrive - Politechnika Śląska\magisterka\pastissmoll\PASTIS_SMOLL\ANNOTATIONS'
num_of_class=18
image_files = [f for f in os.listdir(folder_path) if f.startswith('TARGET_') and f.endswith('.npy')]

height, width = 128, 128
all_unique_colors=[]

unique2=[]
i=0

```

```python
for image_file in image_files:

    immage=np.load(os.path.join(folder_path,image_file))

    image=immage[0,:,:]
    vector_image=image.reshape((-1,1))
    unique_colors, _=np.unique(vector_image,axis=0,return_counts=True)
    unique2.extend(unique_colors)
    for color in vector_image:
        if color.tolist() not in all_unique_colors:
            all_unique_colors.append(color.tolist())

all_unique_colors = np.array(all_unique_colors)
unique_colors_global, _ = np.unique(all_unique_colors, axis=0, return_counts=True)

print(f"Liczba unikalnych kolorów na wszystkich zdjęciach: {len(all_unique_colors)}")
print("Lista unikalnych kolorów:", all_unique_colors)

num=len(unique2)
print('unique=',num)

    
```
